# Qus 2

In Bokeh, glyphs are visual markers used to represent data points on plots. They include shapes like circles, squares,lines , and more, which can be added to a plot display data in different ways. Each glyph corresponds to a specific method in Bokeh's figure object, allowing you to customize the appearance and attribute of the plotted data.
To add glyph to Bokeh plot
1. import necessary modules
2. prepare data
3. create a figure
4. add glyphs
5. customize glyphs
6. display plot


```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
x=[1,2,3,4,5]
y=[6,7,8,9,10]
p=figure(title="Glyphs example",x_axis_label="x-axis",y_axis_label="y-axis")
p.circle(x,y,size=10,color="navy",alpha=0.5,legend_label="Circle")
p.square(x,y,size=12,color="firebrick",alpha=0.8,legend_label="Square")
p.line(x,y,line_width=2,color="green",alpha=0.7,legend_label="Line")
p.title.text_color="olive"
p.title.text_font="times"
p.title.text_font_style="italic"
p.legend.location="top_left"
p.legend.title="Glyphs"


output_notebook()
show(p)
```

```python
from bokeh.plotting import figure,show
from bokeh.io import output_notebook
x=[1,2,3,4,5]
y=[6,7,8,9,10]
p=figure(title="Glyphs example",x_axis_label="x-axis",y_axis_label="y-axis")
p.circle(x,y,size=10,color="blue",alpha=0.5,legend_label="Circle")
p.square(x,y,size=12,color="yellow",alpha=0.8,legend_label="Square")
p.line(x,y,line_width=2,color="green",alpha=0.7,legend_label="Line")
p.title.text_color="olive"
p.title.text_font="helvetica"
p.title.text_font_style="italic"
p.legend.location="top_left"
p.legend.title="Legend"
p.legend.title_text_color="green"

output_notebook()
show(p)
```

# Qus 4

A Bokeh server ia a python server that allows you to create and serve interactive Bokeh plots and applications. It enables you to build rich, interactive data visualization applications that can be accessed and updated dynamically in real time by multiple users.

```python
from bokeh.plotting import figure,curdoc
from bokeh.models import ColumnDataSource
import numpy as np

source=ColumnDataSource(data={"x":[],"y":[]})
p=figure(title="Real-Time",x_axis_label="X",y_axis_label="Y")
line=p.line(x="X",y="Y",source=source)

def update():
  new_data={"x":np.random.random(100),"y":np.random.random(100)}
  source.data=new_data

curdoc().add_periodic_callback(update,1000)
curdoc().add_root(p)
```

# Qus 5


```python
from flask import Flask,render_template
from bokeh.plotting import figure
from bokeh.embed import components
from bokeh.resources import CDN

app = Flask(__name__)

@app.route("/")
def index():
    p=figure(title="My Bokeh Plot",x_axis_label="X",y_axis_label="Y")
    p.circle([1,2,3,4,5],[6,7,2,4,5],size=10,color="navy",alpha=0.5)
    script,div=components(p,CDN)
    return render.template("index.html",script=script,div=div)

if __name__=="__main__":
    app.run("0.0.0.0")
```
